```python
# Q1. What is Ridge Regression, and how does it differ from ordinary least squares regression?

"""Ridge regression is also a linear regression model . However, there is a difference that helps make Ridge regression more 
regularized and thus avoid the problem of overfitting. The OLS model seeks to find the coefficients that minimize the mean squared error. On the other hand, Ridge Regression tries to find the coefficients that minimize the mean squared error and wants the magnitude of coefficients to be as small as possible. That means each feature should have a little effect on the outcome1.

In other words, Ridge regression adds a penalty term to the loss function of OLS regression which shrinks the coefficient 
estimates towards zero. This penalty term is called L2 regularization2.

."""
```

```python
# Q3. How do you select the value of the tuning parameter (lambda) in Ridge Regression?
"""There are several method to find the lambda. 
(1)cross-validation.
(2) grid search. 
(3)L-curve method 

In the L-curve method, we plot the residual sum of squares  against the L2-norm of the coefficients. The optimal 
value of lambda is where the curve has a sharp corner¹.

In cross-validation, we  split your data into training and validation sets. we  then fit our  model on the training set
 and evaluate its performance on the validation set. we  repeat this process for different values of lambda and choose 
 the value that gives you the best performance³.

In grid search, you evaluate your model's performance for different values of lambda and choose the value that gives 
you the best performance¹.

 """
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split


boston = load_diabetes()

X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.3, random_state=42)


ridge = RidgeCV(alphas=[0.1, 1.0, 10.0])


ridge.fit(X_train, y_train)


print("Best lambda: ", ridge.alpha_)
```

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt


diabetes = load_diabetes()


X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.3, random_state=42)


lambdas = np.logspace(-10, 5, 400)


train_errors = np.zeros_like(lambdas)
test_errors = np.zeros_like(lambdas)

for i, lambda_ in enumerate(lambdas):
    
    ridge = Ridge(alpha=lambda_)
    
    ridge.fit(X_train, y_train)

    train_errors[i] = np.mean((y_train - ridge.predict(X_train)) ** 2)

    test_errors[i] = np.mean((y_test - ridge.predict(X_test)) ** 2)


plt.plot(train_errors, test_errors)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Training error')
plt.ylabel('Test error')
plt.title('L-curve for Ridge Regression')
plt.show()

min_idx = np.argmin(test_errors)

print("Best lambda: ", lambdas[min_idx])
```

```python
# Q5. How does the Ridge Regression model perform in the presence of multicollinearity?
"""Ridge Regression performs well in the presence of multicollinearity. Multicollinearity happens when predictor variables
 exhibit a correlation among themselves. Ridge Regression aims at reducing the standard error by adding some bias in the
estimates of the regression. The reduction of the standard error in regression estimates significantly increases the
reliability of the estimates



Ridge regression cab be effective in dealing with multicollinearity ,which is the presence of high correlation 
among the predictor variables. 
 Multicollinearity can lead to unstable and unreliable estimates of the coefficients of the predictor variables,
as small changes in the data can result in large changes in the estimated coefficients.

Ridge Regression addresses multicollinearity by shrinking the estimated coefficients towards zero, effectively reducing
 their variance and making them more stable. This is achieved by adding a penalty term to the least squares objective 
 function that Ridge Regression minimizes, which is proportional to the L2-norm of the coefficients. 

In the presence of multicollinearity, Ridge Regression can help to improve the stability and reliability of the
 coefficient estimates and can also improve the predictive performance of the model. However, it is important
to note that Ridge Regression does not address the root cause of multicollinearity,




"""
```
